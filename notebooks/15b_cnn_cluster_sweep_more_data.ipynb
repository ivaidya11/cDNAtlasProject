```python
import sys, os
sys.path.insert(0, os.path.abspath(os.path.join(os.getcwd(), '..')))

from logic.cnn_logic import cnn_sweep
import numpy as np
import tensorflow as tf
from logic.database import AA_PROPERTIES
import pandas as pd
from logic.clustering_cnn import get_clusters
```

```python
train_raw = pd.read_csv('../data/nanopore_feature_BIG_DATA_cluster.csv')
test_raw = pd.read_csv('../data/randomcontrol_feature_by_cluster.csv')
```

### 4 Cluster CNN

```python
model_4, history_4, X_test_4, y_test_4 = cnn_sweep(
    n_clusters=4, train=train_raw, test=test_raw
    )

n_clusters = 4
```

```python
from collections import defaultdict
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report

y_pred_proba_4 = model_4.predict(X_test_4)           # (n_traces, 20, n_clusters)
y_pred_4 = np.argmax(y_pred_proba_4, axis=-1)      # (n_traces, 20)

y_pred_flat_4 = y_pred_4.flatten()                 # (n_traces * 20,)
y_test_flat_4 = y_test_4.flatten()                 # (n_traces * 20,)

cm = confusion_matrix(y_test_flat_4, y_pred_flat_4)
# cm_norm = cm.astype(float) / cm.sum(axis=1, keepdims=True)

aa_to_cluster = get_clusters(n_clusters)
cluster_to_aas = defaultdict(list)
for aa, cluster in aa_to_cluster.items():
    cluster_to_aas[cluster].append(aa)

# 0-indexed to match model output
tick_labels = [str(sorted(cluster_to_aas[c + 1])) for c in range(n_clusters)]

fig, ax = plt.subplots(1, 1, figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax,
            xticklabels=tick_labels, yticklabels=tick_labels)
ax.set_xlabel('Predicted')
ax.set_ylabel('Actual')
ax.set_title(f'CNN n_clusters={n_clusters} — Counts')
ax.tick_params(axis='x', rotation=45)
ax.tick_params(axis='y', rotation=0)
plt.tight_layout()
plt.show()
print(classification_report(y_test_flat_4, y_pred_flat_4, target_names=tick_labels))


```

```python

def make_tick_labels(n_clusters):
    aa_to_cluster = get_clusters(n_clusters)
    cluster_to_aas = defaultdict(list)
    for aa, cluster in aa_to_cluster.items():
        cluster_to_aas[cluster].append(aa)
    return [str(sorted(cluster_to_aas[c + 1])) for c in range(n_clusters)]

print('------------------------CNN n_clusters=4-------------------------')
print(classification_report(y_test_flat_4, y_pred_flat_4, target_names=make_tick_labels(4)))

print('------------------------CNN n_clusters=5-------------------------')
print(classification_report(y_test_flat_5, y_pred_flat_5, target_names=make_tick_labels(5)))

print('------------------------CNN n_clusters=6-------------------------')
print(classification_report(y_test_flat_6, y_pred_flat_6, target_names=make_tick_labels(6)))

print('------------------------CNN n_clusters=7-------------------------')
print(classification_report(y_test_flat_7, y_pred_flat_7, target_names=make_tick_labels(7)))
```

```python
accuracy_4 = np.mean(y_pred_flat_4 == y_test_flat_4)
accuracy_5 = np.mean(y_pred_flat_5 == y_test_flat_5)
accuracy_6 = np.mean(y_pred_flat_6 == y_test_flat_6)
accuracy_7 = np.mean(y_pred_flat_7 == y_test_flat_7)


plt.figure(figsize=(8, 6))
sns.barplot(x=['4', '5', '6', '7'], y=[accuracy_4, accuracy_5, accuracy_6, accuracy_7], palette='viridis')
plt.xlabel('Number of Clusters')
plt.ylabel('Overall Accuracy')
plt.title('CNN Accuracy by Number of Clusters')
plt.ylim(0, 1)
plt.yticks(np.arange(0, 1.1, 0.05))
plt.tight_layout()
plt.show()
```
